```python
# IMPORT LIBRARIES
```

```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# CREATE FAQ DATASET
```

```python
data = {
    "Question": [
        "What is AI?",
        "What is Machine Learning?",
        "What is Deep Learning?",
        "What is Python?",
        "What is Data Science?"
    ],

    "Answer": [
        "Artificial Intelligence enables machines to perform tasks that normally require human intelligence.",
        "Machine Learning is a subset of AI that learns from data.",
        "Deep Learning is a subset of Machine Learning based on neural networks.",
        "Python is a popular programming language used in AI and Data Science.",
        "Data Science is the process of extracting insights from data."
    ]
}
```

```python
# LOAD FAQ DATA INTO DATAFRAME
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
# DISPLAY DATASET INFORMATION
```

```python
print(df)

print("\nTotal Questions:", len(df))
```

```python
# TEXT PREPROCESSING
```

```python
questions = df["Question"]

print(questions)
```

```python
# APPLY TF-IDF VECTORIZATION
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(
    questions
)

print(tfidf_matrix.shape)
```

```python
# COMPUTE COSINE SIMILARITY
```

```python
similarity_matrix = cosine_similarity(
    tfidf_matrix
)

print(similarity_matrix)
```

```python
# CREATE CHATBOT FUNCTION
```

```python
def chatbot(user_question):

    user_vector = vectorizer.transform(
        [user_question]
    )

    similarity_scores = cosine_similarity(
        user_vector,
        tfidf_matrix
    )

    index = similarity_scores.argmax()

    return df["Answer"][index]
```

```python
# TEST CHATBOT WITH SAMPLE QUERY
```

```python
user_question = "What is AI?"

response = chatbot(user_question)

print("Question:", user_question)
print("Answer:", response)
```

```python
# DISPLAY BEST MATCHING ANSWER
```

```python
user_question = "What is Machine Learning?"

response = chatbot(user_question)

print("Question:", user_question)
print("Answer:", response)
```
